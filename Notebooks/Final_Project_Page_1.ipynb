# **Data Analysis: Gym Member Exercise Database**

The Gym Member Exercise Dataset contains 973 samples of gym data, including key performance indicators such as heart rate, calories burned, and workout duration. Each entry also includes demographic data and experience levels, allowing for comprehensive analysis of fitness patterns, athlete progression, and health trends.

This data will be used to determine which indicators have the largest impact on overall calories burned at the gym.


## Understanding the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('source.csv')

plt.pie(data["Gender"].value_counts(), labels=data["Gender"].value_counts().index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#ff9999'])
plt.title('Gender Distribution in Gym Members')
plt.show()
```

Of the 973 samples included in this dataset, 511 samples were male (52.5%) and 462 samples were female (47.5%).

```python
plt.hist(data['Age'], bins=10, edgecolor='black')
plt.title('Age Distribution of Gym Members')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

The age distribution across the dataset is relatively consistent amongst those between the ages of 38 to 60 with approximately 100 samples per age group. The dataset has a slight increase in those ages 18-22 with an extra 20 samples for this age range. The least represented by a small margin is 22-38 year olds, each bin within this age range being about 20 samples short of the average.

```python
plt.figure(figsize=(10, 6))

sns.countplot(x='Experience_Level', data=data)

plt.title('Distribution of Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.show()
```

We can see in this graph that most of the samples in this dataset have little experience at the gym or a moderate amount of experience at the gym. The least represented group by a large margin are those whose experience levels are higher.

```python
plt.pie(data["Workout_Type"].value_counts(), labels=data["Workout_Type"].value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title('Workout Types Represented')
plt.show()
```

This dataset has an almost even distribution of workout types sampled with a slight preference towards Strength and Cardio training. 

```python
plt.hist(data['Calories_Burned'], bins=20, edgecolor='black')
plt.title('Calories Burned Distribution')
plt.xlabel('Calories Burned')
plt.ylabel('Frequency')
plt.show()
```

On average we can see most samples burned 700-1100 calories in a session with an overall range of 300-1800 caloriesfor all participants.
